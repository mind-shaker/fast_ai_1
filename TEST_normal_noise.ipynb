```python
import numpy as np
import matplotlib.pyplot as plt

# Создаем изображение 3x3 с каждым пикселем, равным 128
original_image = np.full((3, 3), 128, dtype=np.uint8)

# Задаем стандартное отклонение гауссовского шума
sigma = 20

# Генерируем гауссовский шум и добавляем его к оригинальному изображению
gaussian_noise = np.random.normal(0, sigma, original_image.shape)
noisy_image = original_image + gaussian_noise.astype(np.uint8)

# Обрезаем значения, чтобы они оставались в пределах 0-255
noisy_image = np.clip(noisy_image, 0, 255)

# Выводим результаты
print("Original Image:")
print(original_image)
print("\nGaussian Noise Matrix:")
print(gaussian_noise)
print("\nNoisy Image:")
print(noisy_image)

# Отобразим изображение с шумом
plt.imshow(noisy_image, cmap='gray', vmin=0, vmax=255)
plt.title("Noisy Image")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Задаем параметры шума
duration = 1.0  # Длительность шума в секундах
sampling_rate = 1000  # Частота дискретизации (количество измерений в секунду)

# Генерация белого шума
t = np.linspace(0, duration, int(sampling_rate * duration), endpoint=False)
white_noise = np.random.normal(0, 1, len(t))  # Генерация нормально распределенного шума

# Визуализация шума
plt.plot(t, white_noise)
plt.title('White Noise')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Задаем параметры шума
duration = 1.0  # Длительность шума в секундах
sampling_rate = 200  # Частота дискретизации (количество измерений в секунду)

# Генерация белого шума
t = np.linspace(0, duration, int(sampling_rate * duration), endpoint=False)
white_noise = np.random.normal(0, 1, len(t))  # Генерация нормально распределенного шума

# Визуализация шума
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(t, white_noise)
plt.title('White Noise')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')

# Визуализация массива значений
plt.subplot(1, 2, 2)
plt.hist(white_noise, bins=50, density=True, alpha=0.75, color='blue')
plt.title('Histogram of White Noise Values')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.tight_layout()
plt.show()
```
